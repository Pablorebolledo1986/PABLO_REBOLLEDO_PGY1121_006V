```python
import random

def FormatRut(rut):
    Clean = ''.join(filter(str.isalnum, rut))
    Verificador = Clean[-1]
    Digitos = Clean[:-1]
    Digitos_Invertidos = Digitos[::-1]
    Digitos_Formateados = ".".join([Digitos_Invertidos[i:i+3] for i in range(0, len(Digitos_Invertidos), 3)])[::-1]
    return Digitos_Formateados.upper() + "-" + Verificador

def CheckRUT(rut):
    def Calcular_Verificador(digitos):
        factor = 2
        suma = 0

        for digito in digitos[::-1]:
            suma += int(digito)*factor
            factor += 1
            if factor > 7:
                factor = 2
        Digito_Verificador = 11-(suma%11)
        if Digito_Verificador == 11:
            Digito_Verificador = "0"
        elif Digito_Verificador == 10:
            Digito_Verificador = "K"
        return str(Digito_Verificador)

    Clean = ''.join(filter(str.isalnum, rut))
    D = Clean[:-1]
    V = Clean[-1]
    if (Clean[:-1].isdigit()) and (V.isdigit() or V.lower() == "k") and (len(D) >= 7 and len(D) < 10) and (Calcular_Verificador(D) == V):
        return True
    else:
        return False

def grabar_persona():
    print("Grabar persona")
    print("------------------------")
    Rut = input("RUT: ")
    while not CheckRUT(Rut):
        print("RUT inválido. Ingrese un RUT válido.")
        Rut = input("RUT: ")
    Nombre = input("Nombre: ")
    while len(Nombre) < 8:
        print("Nombre debe tener al menos 8 caracteres ")
        Nombre = input("Nombre: ")
    Edad = input("Edad: ")
    while not Edad.isdigit() or int(Edad) < 0:
        print("Edad inválida. Ingrese una edad válida.")
        Edad = input("Edad: ")
    Pais_Nacimiento = input("País de nacimiento: ")
    Ciudad_Nacimiento = input("Ciudad de nacimiento: ")
    print("\nDatos grabados")
    print("------------------------")
    print(f"RUT: {FormatRut(Rut)}")
    print(f"Nombre: {Nombre}")
    print(f"Edad: {Edad} años")
    if Pais_Nacimiento.lower() == "chile":
        print("País de nacimiento: Chile")
    else:
        print("País de nacimiento: Extrajero")
        print(f"Ciudad de nacimiento: {Ciudad_Nacimiento}")
    print("------------------------\n")

def buscar_persona():
    print("Buscar persona")
    print("------------------------")
    Rut = input("Ingrese el RUT de la persona a buscar: ")
    print("\nDatos de la persona buscada")
    print("------------------------")
    print(f"RUT: {FormatRut(Rut)}")
    print("------------------------\n")

def imprimir_certificados():
    print("Imprimir certificados")
    print("------------------------")
    nombre = input("Ingrese el nombre de la persona a imprimir el certificado: ")
    rut = input("Ingrese el RUT de la persona: ")
    print("------------------------")
    print("Certificado de Nacimiento")
    print("------------------------")
    print(f"Circunscripción: {random.choice(['Santiago', 'Valparaíso', 'Concepción'])}")
    print(f"Nro. inscripción: {random.randint(1000, 9999)}")
    print(f"Nombre inscrito: {nombre}")
    print(f"RUT: {FormatRut(rut)}")
    print(f"Fecha de nacimiento: {random.randint(1, 28)}/{random.randint(1, 12)}/{random.randint(1950, 2020)}")
    print(f"Hora de nacimiento: {random.randint(0, 23)}:{random.randint(0, 59)}")
    print(f"Sexo: {random.choice(['femenino', 'masculino'])}")
    print(f"Nombre del padre: {random.choice(['Juan', 'Pedro', 'Carlos'])}")
    print(f"Nombre de la madre: {random.choice(['María', 'Ana', 'Laura'])}")
    print(f"Fecha y año de registro: {random.randint(1, 28)}/{random.randint(1, 12)}/{random.randint(2020, 2022)}")
    print("------------------------\n")

def eliminar_persona():
    print("Eliminar persona")
    print("------------------------")
    Rut = input("Ingrese el RUT de la persona a eliminar: ")
    while not CheckRUT(Rut):
        print("RUT inválido. Ingrese un RUT válido.")
        Rut = input("RUT: ")

    print("La persona ha sido eliminada exitosamente.")
    print("------------------------\n")

while True:
    print("Sistema de Registro Civil")
    print("------------------------")
    print("1. Grabar persona")
    print("2. Buscar persona")
    print("3. Imprimir certificados")
    print("4. Eliminar persona")
    print("5. Salir")
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        grabar_persona()
    elif opcion == "2":
        buscar_persona()
    elif opcion == "3":
        imprimir_certificados()
    elif opcion == "4":
        eliminar_persona()
    elif opcion == "5":
        print("Saliendo del programa...")
        break
    else:
        print("Opción inválida. Seleccione una opción válida.\n")
print("**************************************************************************")
print("Gracias por usar el totem de auto servicio del registro civil hasta pronto")
print("**************************************************************************")
```
